```python
# ==============================================================================
# TASK 2: EXPLORATORY DATA ANALYSIS (EDA)
# Description: Automatically explore data structure, trends, and anomalies.
# Author: [Arpana kumari]
# ==============================================================================

# 1. Necessary Libraries Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style for clean GitHub-ready visualizations
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("--- Step 1: Loading Dataset & Asking Initial Questions ---")
# Loading a built-in dataset from seaborn for instant execution
# (You can replace this with: df = pd.read_csv('your_file.csv') if you have your own data)
df = sns.load_dataset('iris')
print("Sample Dataset 'Iris' loaded successfully!")
print(f"Initial Question: What is the distribution of features across different species?\n")


print("--- Step 2: Exploring Data Structure ---")
# Checking shape, columns, and data types
print(f"Dataset Shape: {df.shape} (Rows, Columns)")
print("\n--- Data Info & Types ---")
print(df.info())
print("\n--- First 5 Rows of the Dataset ---")
print(df.head())


print("\n--- Step 3 & 4: Descriptive Statistics & Identifying Trends/Anomalies ---")
# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Identifying missing values (Data Issues)
print("\nChecking for Missing Values:")
print(df.isnull().sum())

# Detecting potential outliers/anomalies using Boxplot
print("\nGenerating Visualizations...")

# Plot 1: Boxplot to check for Outliers/Anomalies
plt.figure(figsize=(10, 5))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.title('Boxplot for Outlier & Anomaly Detection', fontsize=14, fontweight='bold')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()

# Plot 2: Pairplot to observe trends, patterns, and test hypotheses
# Hypothesis: Different species have distinct cluster patterns based on feature measurements.
print("\nGenerating Pairplot to identify patterns and relationships...")
sns.pairplot(df, hue='species', palette='Set2')
plt.suptitle('Pairplot: Trends and Relationships between Variables', y=1.02, fontsize=14, fontweight='bold')
plt.show()


print("\n--- Step 5: EDA Summary & Potential Issues Detected ---")
missing_count = df.isnull().sum().sum()
if missing_count == 0:
    print("✔️ No missing values detected. Data is clean for further analysis.")
else:
    print(f"⚠️ Warning: Found {missing_count} missing values. Needs cleaning in next phase.")

print("✔️ EDA Task completed successfully! Code is structured and documented for GitHub.")
```
